```python
import nltk 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

I wanted to take a look at some of the data in the child-directed language database, CHILDES. Luckily, NLTK has a corpus reader designed to analyze data from CHILDES. I used examples from the CUNY site on using CHILDES with NLTK as a starting place for this assignment.

The CHILDES corpus isn't included in the base NLTK corpora set, so I downloaded a small subset of the North American English dataset, particularly the Tardif set: 

http://childes.talkbank.org/access/Eng-NA/Tardif.html

Twila Tardif has done a lot of interesting work on development of vocabulary across cultures, and this dataset was created from recodings of the English monolingual sample in a study examining Chinese and English lexical development.

I downloaded the unzipped file to the root folder that NLTK expects: User/nltk_data/corpora/chiles/data-xml/Eng-USA-MOR (note: the folder was created when I downloaded some other NLTK corpora) 

```python
# import CHILDES corpus reader
from nltk.corpus.reader import CHILDESCorpusReader

# set root folder
corpus_root = nltk.data.find('corpora/childes/data-xml/Eng-USA-MOR/')

# read in Tardif dataset
tardif = CHILDESCorpusReader(corpus_root,'Tardif/.*.xml')
```

You can access data from the CHILDES database tagged by speaker (e.g., mother, child, father), and also tagged by part of speech, which is cool. But this kid's data does not look very interesting.

```python
# get list of words used by speaker (child) in the first Tardif file, tagged by part of speech (relation)
tardif.words('Tardif/e01.xml',speaker=['CHI'], relation=True)
```

The CHILDES data is also tagged with the speaker's age, which is obviously useful for developmental work. This comes in a funky format, but can be converted to months which is easier to interpret.

```python
# get age in months for each Tardif file
age = tardif.age(month=True)
print(age)
```

One standard measure in corpus analyses is mean length of utterance (MLU), which is the average number of words used per utterance. I would expect this to increase over the first few years, then plateau (our sentences don't just keep getting longer and longer).  

```python
# get MLU for each Tardif file
MLU = tardif.MLU(speaker=['CHI'])
print(MLU)
```

It might be interesting to take a look at whether this expected pattern is visible in this dataset by plotting MLU by age.

```python
x = age
y = MLU

#create a figure 
fig, ax = plt.subplots()

#create a trendline
fit = np.polyfit(x, y, deg=1)
ax.plot(x, fit[0] * x + fit[1], color='red')
ax.scatter(x, y)
ax.set_xlabel('Age (in Month)')
ax.set_ylabel('MLU')
```

No so much. This might be in part due to the narrow age range (1.5-2yrs). It's possible that with more data this trend would look a little better. So I downloaded all of the CHILDES English corpus data, and ran through the same process. This took forever to run, but I guess that makes sense since it's going through 6,000+ files? 

```python
childes = CHILDESCorpusReader(corpus_root,u'.*.xml')
# initialize age and MLU variables
age = []
MLU = []

# for loop to get age and MLU for each file in the CHILDES corpus
for filename in childes.fileids():
    age.extend(childes.age(filename, month = True))
    MLU.extend(childes.MLU(filename))
```

```python
df = pd.DataFrame({"Age": age, "MLU" : MLU})
x = df.Age
y = df.MLU

# create a figure 
fig, ax = plt.subplots(figsize=(50,20))

#create a trendline
temp = np.isfinite(x) & np.isfinite(y)
fit = np.polyfit(x[temp], y[temp], deg=1)
ax.plot(x, fit[0] * x + fit[1], color='red')
ax.scatter(x, y, marker = 'o')
ax.set_xlabel('Age (in Months)', fontsize=30)
ax.set_ylabel('MLU', fontsize=30)
ax.tick_params(labelsize=30)
```

Cool, the expected trend does exist! I wanted to draw curve with a 2nd degree polyinomial here, but couldn't figure out a way to do this that worked. I would expect the MLU average to plateau around 5-10 word, which appears to happen around 36mths, when we expect children to start speaking in relatively complete sentences. 

On a related note, what are the most common words early (under 12mth) vs late (around 3yrs) in language development? This also took quite a while to run. Is there a more efficient way to do something like this or the above? Or is this just inevitable given the number of files?

```python
words12 = []
words36 = []
for filename in childes.fileids():
    age = childes.age(filename, month = True)[0]
    if age is None:
        pass
    elif age < 12:
        words12.extend(childes.words(filename,speaker=["CHI"]))
    elif age > 36:
        words36.extend(childes.words(filename,speaker=["CHI"]))
counts12 = nltk.FreqDist(words12)
counts36 = nltk.FreqDist(words36)
```

```python
counts12.plot(15, title = 'Most Frequent Words Under 12mths')
counts36.plot(15, title = 'Most Frequent Words Over 36mths')
```

Well, there is definitely a difference in the most common words used by infants under 1 year and toddlers over age 3. Young babies utterances mostly consist of babbling, whereas older children show evidence of more complex speech (e.g. use of pronouns, determiners, conjunctions, etc. like adults).
