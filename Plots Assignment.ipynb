One way to look at this birth data is by time of year. Are more people born 9 months after the cold fall and winter days than following warmer months, as everyone expects?

```python
# import necessary packages
import pandas as pd
import matplotlib as plt
import seaborn as sns
%matplotlib inline
```

```python
# read in data looing at the number of births per month/day/day of week per during each year 1994-2003
birth_data = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/births/US_births_1994-2003_CDC_NCHS.csv')
```

```python
birth_data.head()
```

```python
# bar graph displaying mean births grouped by month (1 = January)
ax = birth_data.groupby(['month'])['births'].mean().plot.pie(
    title = 'Mean Proportion Births by Month 1994-2003', # add title
    colormap = 'RdYlGn',  # change color
    figsize = [5,5]); # change size
```

From the pie chart it looks like each month takes a pretty even share of the births. Could look at specific numbers though using a bar graph. 

```python
# bar graph displaying mean births grouped by month
ax = birth_data.groupby(['month'])['births'].mean().plot.bar(
    title = 'Avg Births by Month 1994-2003', # add title
    colormap = 'summer', # change color
    ylim = [8000,13000]) # zoom to appropriate y axis dimension
ax.set_ylabel('Births') # set axis labels
ax.set_xlabel('Month')
xlab = ['January', 'February', 'March', 'April','May','June','July','August','September','October','November','December']
ax.set_xticklabels(xlab); # change numeric labels to words
```

From the bar graph it looks like there is the expected bump in births during the summer months. September has the clear baby bump, maybe because of the cold winter months or December holidays?
